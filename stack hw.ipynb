```python
class Stack:
    def __init__(self):
        self.stack = []
    
    def push(self, item):
        self.stack.append(item)
    
    def pop(self):
        if not self.is_empty():
            return self.stack.pop()
        return None
    
    def peek(self):
        if not self.is_empty():
            return self.stack[-1]
        return None
    
    def is_empty(self):
        return len(self.stack) == 0
    
    def size(self):
        return len(self.stack)
    
    def print_stack(self):
        print(self.stack)

    def reverse(self):
        if not self.is_empty():
            temp = self.pop()
            self.reverse()
            self.insert_at_bottom(temp)
    
    def insert_at_bottom(self, item):
        if self.is_empty():
            self.push(item)
        else:
            temp = self.pop()
            self.insert_at_bottom(item)
            self.push(temp)

    def get_min(self):
        if self.is_empty():
            return None
        min_val = self.stack[0]
        for item in self.stack:
            if item < min_val:
                min_val = item
        return min_val
    
    def sort_descending(self):
        temp_stack = Stack()
        while not self.is_empty():
            temp = self.pop()
            while not temp_stack.is_empty() and temp_stack.peek() < temp:
                self.push(temp_stack.pop())
            temp_stack.push(temp)
        
        while not temp_stack.is_empty():
            self.push(temp_stack.pop())

stack = Stack()
stack.push(3)
stack.push(1)
stack.push(4)
stack.push(2)

print("Original Stack:")
stack.print_stack()

stack.reverse()
print("Reversed Stack:")
stack.print_stack()

print("Minimum Element:", stack.get_min())

stack.sort_descending()
print("Sorted Stack in Descending Order:")
stack.print_stack()
```
